```python
# Linear Search
def linear_search(arr, target):
    for i in range(len(arr)):      # loop semua elemen (dari awal sampai akhir)
        if arr[i] == target:       # cek apakah elemen sama dengan target
            return i               # kalau ketemu, kembalikan index
    return -1                      # kalau tidak ketemu, return -1

# contoh
data = [10, 20, 30, 40, 50]
print(linear_search(data, 30))  # output: 2
# Kompleksitas:
# Best  : O(1)  (ketemu di awal)
# Worst : O(n)  (cek semua data)
```

```python
# Binary Search (Iteratif)
def binary_search(arr, target):
    left, right = 0, len(arr) - 1   # batas kiri & kanan array

    while left <= right:            # selama masih ada range pencarian
        mid = (left + right) // 2   # cari tengah

        if arr[mid] == target:      # kalau ketemu
            return mid
        elif arr[mid] < target:     # kalau target lebih besar
            left = mid + 1          # geser ke kanan
        else:
            right = mid - 1         # geser ke kiri

    return -1                       # tidak ditemukan

# contoh (data harus terurut)
data = [10, 20, 30, 40, 50]
print(binary_search(data, 40))  # output: 3
# Kompleksitas:
# Best  : O(1)
# Worst : O(log n)
```

```python
# Binary Search (Rekursif)
def binary_search_recursive(arr, left, right, target):
    if left > right:                # kondisi berhenti
        return -1

    mid = (left + right) // 2       # titik tengah

    if arr[mid] == target:          # kalau ketemu
        return mid
    elif arr[mid] < target:         # ke kanan
        return binary_search_recursive(arr, mid + 1, right, target)
    else:                           # ke kiri
        return binary_search_recursive(arr, left, mid - 1, target)

# contoh
data = [10, 20, 30, 40, 50]
print(binary_search_recursive(data, 0, len(data)-1, 20))  # output: 1
# Kompleksitas:
# Best  : O(1)
# Worst : O(log n)
```

```python
# Jump Search
import math

def jump_search(arr, target):
    n = len(arr)                        # jumlah data
    step = int(math.sqrt(n))            # ukuran lompatan (akar n)
    prev = 0

    # lompat blok
    while arr[min(step, n) - 1] < target:  # cek ujung blok
        prev = step                       # pindah blok
        step += int(math.sqrt(n))         # tambah lompatan
        if prev >= n:                     # kalau lewat batas
            return -1

    # linear search di blok
    for i in range(prev, min(step, n)):   # cari di blok
        if arr[i] == target:
            return i

    return -1

# contoh
data = [10, 20, 30, 40, 50, 60, 70]
print(jump_search(data, 60))  # output: 5
# Kompleksitas:
# Best  : O(1)
# Worst : O(√n)
```

```python
# Interpolation Search
def interpolation_search(arr, target):
    low = 0
    high = len(arr) - 1

    while low <= high and target >= arr[low] and target <= arr[high]:
        if low == high:                 # jika tinggal 1 elemen
            if arr[low] == target:
                return low
            return -1

        # rumus interpolasi (perkiraan posisi)
        pos = low + ((target - arr[low]) * (high - low) //
                     (arr[high] - arr[low]))

        if arr[pos] == target:
            return pos
        elif arr[pos] < target:         # geser ke kanan
            low = pos + 1
        else:                           # geser ke kiri
            high = pos - 1

    return -1

# contoh
data = [10, 20, 30, 40, 50]
print(interpolation_search(data, 50))  # output: 4
# Kompleksitas:
# Best  : O(1)
# Average : O(log log n)
# Worst : O(n)
```
