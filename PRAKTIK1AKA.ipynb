```python
def total(angka, n):
    jumlah = 0
    for i in range(n):
        jumlah = jumlah + angka[i]
    return jumlah

angka = [4,6,8,10]
n = len(angka)
print(total(angka, n))
```

```python
a = 3
b = 3
angka = [(2,4,6),(1,3,5),(7,8,9)]
total = 0

for i in range(a):
    for j in range(b):
        total += angka[i][j]

print(total)
```
